```python
import numpy as np
import matplotlib.pyplot as plt

Xi = np.array([-4, -3, -1, 3])
Fi = np.array([-15, 5, 3, -1])

X_values = np.array([-3.5, -2, 1.5, 2])

def lagrange_polynomial(Xi, Fi, x):
    n = len(Xi)
    result = 0
    for i in range(n):
        term = Fi[i]
        for j in range(n):
            if i != j:
                term *= (x - Xi[j]) / (Xi[i] - Xi[j])
        result += term
    return result

F_values = np.array([lagrange_polynomial(Xi, Fi, x) for x in X_values])
print("Наближені значення функції в заданих точках:", F_values)

x_plot = np.linspace(min(Xi)-1, max(Xi)+1, 500)
y_plot = [lagrange_polynomial(Xi, Fi, x) for x in x_plot]

plt.plot(Xi, Fi, 'o', label="Вихідні точки")
plt.plot(x_plot, y_plot, label="Інтерполяційний багаточлен Лагранжа")
plt.plot(X_values, F_values, 'x', label="Обчислені точки")
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Інтерполяційний багаточлен Лагранжа')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

Xi = np.array([-4, -3, -1, 3])
Fi = np.array([-15, 5, 3, -1])

X_values = np.array([-3.5, -2, 1.5, 2])

poly = lagrange(Xi, Fi)

F_values = poly(X_values)

x_new = np.linspace(min(Xi)-1, max(Xi)+1, 500)
y_new = poly(x_new)

plt.figure(figsize=(7, 5))
plt.plot(x_new, y_new, 'b-', label="Інтерполяційний багаточлен Лагранжа")
plt.plot(Xi, Fi, 'ro', label="Вихідні точки")
plt.plot(X_values, F_values, 'go', label="Обчислені точки")

plt.plot(X_values[0], F_values[0], 'bo', label="f(-3.5)")
plt.plot(X_values[1], F_values[1], 'co', label="f(-2)")
plt.plot(X_values[2], F_values[2], 'mo', label="f(1.5)")
plt.plot(X_values[3], F_values[3], 'yo', label="f(2)")

plt.title("Інтерполяційний багаточлен Лагранжа з точками перевірки")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
