Розв’язати систему нелінійних рівнянь з точністю до 0.001. Відокремлення коренів виконати графічно. Уточнення коренів провести методом простої ітерації.

```python
import numpy as np
import matplotlib.pyplot as plt

x_min, x_max = -2, 4
y_min, y_max = -3, 3
step = 0.01

x, y = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))

eq1 = np.sin(y + 0.5) - x - 1
eq2 = np.cos(x - 2) + y

fig, ax = plt.subplots(figsize=(7, 7))

ax.contour(x, y, eq1, levels=[0], colors='red', label='sin(y + 0.5) - x - 1')

ax.contour(x, y, eq2, levels=[0], colors='blue', label='cos(x - 2) + y')

ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Графік системи рівнянь')

ax.arrow(x_min, 0, x_max - x_min, 0, head_width=0.1, head_length=0.1, fc='black', ec='black')
ax.arrow(0, y_min, 0, y_max - y_min, head_width=0.1, head_length=0.1, fc='black', ec='black')

plt.grid(True)

plt.show()
```

```python
import numpy as np
from scipy import optimize

x0 = 1.5
y0 = -0.5

def f1(y):
    return np.sin(y + 0.5) - 1

def f2(x):
    return -np.cos(x - 2)

def iter(x, y, e):
    xn = x
    yn = y
    xn1 = f1(yn)
    yn1 = f2(xn)
    n = 1
    while (abs(xn1 - xn) >= e) or (abs(yn1 - yn) >= e):
        xn = xn1
        yn = yn1
        xn1 = f1(yn)
        yn1 = f2(xn)
        n += 1
    print('Simple iteration:')
    print('x =', xn1, '\ny =', yn1, '\nThe number of iterations =', n)

iter(x0, y0, 0.001)

def f3(x):
    return np.sin(x[1] + 0.5) - x[0] - 1, np.cos(x[0] - 2) + x[1]

s = optimize.root(f3, [x0, y0], method='hybr')

print('Check:', s.x)
```
