```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton, bisect

def f(x):
    return 3*x**4 - 10*x**3 + x**2 - 5*x - 3

def df(x):
    return 12*x**3 - 30*x**2 + 2*x - 5

def find_intervals(f, start, end, step):
    intervals = []
    x = start
    while x < end:
        if f(x) * f(x + step) < 0:
            intervals.append((x, x + step))
        x += step
    return intervals

start = -2
end = 3
step = 0.1

intervals = find_intervals(f, start, end, step)

if intervals:
    print("Функція змінює знак на наступних інтервалах:")
    for i, interval in enumerate(intervals):
        print(f"Інтервал {i+1}: {interval}")
else:
    print("Корені не знайдено на даному інтервалі.")

initial_guess = 2.0
root_newton = newton(f, initial_guess, fprime=df)
print("Корінь методом Ньютона:", root_newton)

if intervals:
    a, b = intervals[0]
    root_combined = bisect(f, a, b)
    print("Корінь комбінованим методом:", root_combined)
else:
    print("Не вдалося знайти інтервал для методу бісекції.")


x = np.linspace(-1, 3, 400)
y = f(x)

plt.plot(x, y, label='3x^4 - 10x^3 + x^2 - 5x - 3')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(root_newton, color='red', linestyle='--', label=f'Ньютон: x={root_newton:.4f}')
if intervals:
    plt.axvline(root_combined, color='blue', linestyle='--', label=f'Комбінований: x={root_combined:.4f}')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Корені нелінійного рівняння')
plt.grid(True)
plt.show()
```
